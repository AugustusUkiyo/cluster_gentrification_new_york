# Import Packages

```python
%matplotlib inline
import pandas as pd
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
import pickle
import numpy as np
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
sns.set(rc={'figure.figsize':(15,10)})
plt.style.use('fivethirtyeight')
sns.set_palette(sns.diverging_palette(250,700, s= 70 , l= 10, n= 12))
```

```python
df = pd.read_pickle('PKL/nyc_full_pc.pkl')
df2 = pd.read_pickle('PKL/nyc_full.pkl')
```

```python
df.drop(df.index[0],inplace = True)
df2.drop(df2.index[0],inplace = True)
```

```python
df.describe()
```

```python
ax = sns.distplot(df2['POP10_POP00'],bins=300)

ax.set(xlabel='Population Change 2000-2010',ylabel='Frequency')
ax.set_title('Distribution of Population Changes Across Tracts')
plt.savefig('Images/Population_change_dist.png', dpi=300)
```

```python

ax = sns.barplot(y='POP10_POP00',x='TRACTID',data=df[df['POP10_POP00']<-25])

ax.set(xlabel='Tract ID',ylabel='Change in Population Percentage (2000-2010)')
ax.set_title('Census Tracts with Largest Population Declines')
# plt.savefig('Population_change_negative.png', dpi=300)
```

```python
fig, ax = plt.subplots(figsize = (15,10))
sns.barplot(y='POP10_POP00',x='TRACTID',data=df[df['POP10_POP00']>110])

ax.set(xlabel='Tract ID',ylabel='Change in Population Percentage (2000-2010)')
ax.set_title('Census Tracts with Largest Population Increases')

plt.xticks(rotation=80)
# plt.savefig('Population_change_negative.png', dpi=300)
```

```python
ax = sns.boxplot(data=df[['NHWHT10%_NHWHT00%','NONWHT10%_NONWHT00%']])

ax.set_title('Change in White and Non-white Population Changes (2000-2010) by Race')
# plt.savefig('Race_pop_changes.png', dpi=300)
```

```python
fig, ax = plt.subplots(figsize = (15,10))
sns.distplot(df2['NHWHT10%_NHWHT00%'],bins=300)
sns.distplot(df2['NONWHT10%_NONWHT00%'],bins=300)

ax.set(xlabel='Population Change 2000-2010',ylabel='Frequency', ylim = (0,0.02), xlim = (-100,1000))
ax.set_title('Distribution of Population Changes Across Tracts')

plt.savefig('Images/Population_change_dist.png', dpi=300)
```
