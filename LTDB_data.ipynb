## Split the features into categories

Create a function that will take care of the different types of features and turn them into percentages out of the total of each category

```python
ppl_cols = list(nyc_00.columns.values[9:-4]) # all of the personal categories
house_cols = list(nyc_00.columns.values[-4:-2])# all of the housing categories
```

```python
def pipeline(df,ppl_list,house_list):
    for x in df.columns.values:
        if x in ppl_list:
            df[x + '%'] = 100 * df[x] / df['POP00']
        elif x in house_list:
            df[x + '%'] = 100 * df[x] / df['HU00']
    # the percentage of non 
    df['NONWHT00%'] = 100*(df.iloc[:,10:14].sum(1)/df['POP00'])
    df.drop(columns = ['HU00%'], axis = 1, inplace = True)
    df.drop(df.columns[9:51],axis=1,inplace=True)
    df.drop(df.columns[4:8],axis=1,inplace=True)

```

```python
nyc_00p = nyc_00.copy()

pipeline(nyc_00p,ppl_cols,house_cols)
nyc_00p.head()
```

## Filter out non-residential areas.

Census tracts with small amounts of people will be dropped. These include parks, cemeteries, and heavily commercial areas. 

Make the bar to have at least 700 people in an observation. 

```python
nyc_00p = nyc_00p[nyc_00p['POP00'].gt(700)]
print(nyc_00p.shape)
nyc_00p.head()
```

## We must fill in the NaN values

We only have two NaNs left, those will just be replaced with 0's. We will also rename the Tract ID column to make it easier to remember for future data frames. 

```python
nyc_00p['OWN00%'] = nyc_00p['OWN00%'].fillna(0)
nyc_00p['RENT00%'] = nyc_00p['RENT00%'].fillna(0)
nyc_00p.rename(columns={'TRTID10':'TRACTID'},inplace=True)

nyc_00p.isna().sum()
```

## Pairing matching begins with the list comprehensions below

Each list will pull only the features that have the same year that they are collecting

```python
lst_10 = [x for x in nyc_both.columns.values if '10' in x]
lst_00 = [x for x in nyc_both.columns.values if '00' in x]
lst = list(zip(lst_10,lst_00))
lst
```

## In the for loop below, we begin the process of taking the percent change

Also in the code below is a way to fill in infinity and NaNs that come up during the calculations. Some variables are 0 and will output infinity in this equation.

```python
for i, x in zip(lst_10,lst_00):
    try:
        nyc_both[i +'_'+ x] = 100 * ((nyc_both[i] - nyc_both[x])/ (nyc_both[x]))
    except ZeroDivisionError:
        0

nyc_both = nyc_both.replace([np.inf, -np.inf], np.nan)       
nyc_both = nyc_both.fillna(0)    
nyc_both.head()

```
