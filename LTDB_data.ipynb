# Read the 2000 Full count csv, take a look, and isolate NYC

```python
df_00 = pd.read_csv('DATA/LTDB_Std_All_fullcount/LTDB_Std_2000_fullcount.csv',sep=',', engine='python')      
print(df_00.shape)
print(df_00[df_00.state.eq('NY')].county.value_counts().head())
df_00.head()
```

```python
nyc_00 = df_00[ (df_00['state'].eq('NY'))]
nyc_00 = nyc_00[    (nyc_00['county'].eq('Bronx County')) 
                  | (nyc_00['county'].eq('Kings County')) 
                  | (nyc_00['county'].eq('Queens County')) 
                  | (nyc_00['county'].eq('New York County'))]
print(nyc_00.shape)
nyc_00.head()

```

## Reworking the data so that the percentage of each population category is shown.

The percentages will be much more meaningful than the total counts, especially as population increases overall.

```python
nyc_00.columns.values
```

```python
ppl_cols = list(nyc_00.columns.values[9:-4])
house_cols = list(nyc_00.columns.values[-4:-2])
```

```python
def pipeline(df,ppl_list,house_list):
    for x in df.columns.values:
        if x in ppl_list:
            df[x + '%'] = 100 * df[x] / df['POP00']
        elif x in house_list:
            df[x + '%'] = 100 * df[x] / df['HU00']
    # the percentage of non 
    df['NONWHT00%'] = 100*(df.iloc[:,10:14].sum(1)/df['POP00'])
    df.drop(columns = ['HU00%'], axis = 1, inplace = True)
    df.drop(df.columns[9:51],axis=1,inplace=True)
    df.drop(df.columns[4:8],axis=1,inplace=True)

```

```python
nyc_00p = nyc_00.copy()

pipeline(nyc_00p,ppl_cols,house_cols)
nyc_00p.head()
```

## Census tracts with small amounts of people will be dropped. Not an accurate representation

```python
print(nyc_00p.POP00.median())
print(nyc_00p.POP00.mean())
```

```python
nyc_00p = nyc_00p[nyc_00p['POP00'].gt(700)]
print(nyc_00p.shape)
nyc_00p.head()
```

```python
nyc_00p.isna().sum()
```

### We only have two NaNs left, those will just be replaced with 0's.

```python
nyc_00p['OWN00%'] = nyc_00p['OWN00%'].fillna(0)
nyc_00p['RENT00%'] = nyc_00p['RENT00%'].fillna(0)
nyc_00p.isna().sum()
```

```python
nyc_00p.to_pickle('PKL/nyc_00p.pkl')
```
