## Reworking the data so that the percentage of each population category is shown.

The percentages will be much more meaningful than the total counts, especially as population increases overall.

```python
nyc_00.columns.values
```

```python
ppl_cols = list(nyc_00.columns.values[9:-4])
house_cols = list(nyc_00.columns.values[-4:-2])
```

```python
def pipeline(df,ppl_list,house_list):
    for x in df.columns.values:
        if x in ppl_list:
            df[x + '%'] = 100 * df[x] / df['POP00']
        elif x in house_list:
            df[x + '%'] = 100 * df[x] / df['HU00']
    # the percentage of non 
    df['NONWHT00%'] = 100*(df.iloc[:,10:14].sum(1)/df['POP00'])
    df.drop(columns = ['HU00%'], axis = 1, inplace = True)
    df.drop(df.columns[9:51],axis=1,inplace=True)
    df.drop(df.columns[4:8],axis=1,inplace=True)

```

```python
nyc_00p = nyc_00.copy()

pipeline(nyc_00p,ppl_cols,house_cols)
nyc_00p.head()
```

```python
nyc_00p.isna().sum()
```

```python
nyc_00p = nyc_00p[nyc_00p['POP00'].gt(700)]
print(nyc_00p.shape)
nyc_00p.head()
```

```python
nyc_00p.isna().sum()
```
